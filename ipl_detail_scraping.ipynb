```python
import pandas as pd
import time
from time import sleep
from selenium import webdriver
from selenium.webdriver.common.by import By
import csv
from slugify import slugify
from selenium.webdriver.common.action_chains import ActionChains
import math
from slugify import slugify
```

```python
def ipl_details(FullTeamName):
    driver = webdriver.Chrome(executable_path=r'C:\Users\F1\Downloads\chromedriver_win32\chromedriver.exe')

    slugify(FullTeamName, separator="-")
    driver.get(f'https://www.iplt20.com/teams/{slugify(FullTeamName, separator="-")}')
    driver.maximize_window()
    time.sleep(1)

    try:
        cookie = driver.find_element(
            By.XPATH,
            '//button[@class="cookie__accept btn btn-primary"]'
        )
        cookie.click()
    except:
        pass

    squad_btn = driver.find_element(
        By.XPATH, 
        '/html/body/div[1]/div[3]/a[4]'
    )
    squad_btn.click()

    players = driver.find_elements(
        By.XPATH, 
        "//ul/li[@class='dys-box-color ih-pcard1']"
    )

    profile_link=[]
    profiles = driver.find_elements(
        By.XPATH, 
        "//div[@class='ih-vp-btn']/a"
    )
    for profile in profiles:
        profile_link.append(profile.get_attribute("href"))

    players_details = []
    for i, player in enumerate(players):

        driver.get(profile_link[i])

        lst = []

        img_elem = driver.find_element(
            By.XPATH, 
            "//div[@class='ih-td-bimg']/img"
        )
        img = img_elem.get_attribute("src")
        lst.append(img)

        name = driver.find_element(
            By.XPATH, 
            "//div[@class='ih-td-n']"
        )
        lst.append(name.text)

        lefts = driver.find_elements(
            By.XPATH, 
            "//div[@class='bbot-l']/h2"
        )
        for left in lefts:
            x = left.text
            lst.append(x.split(': ')[1])

        rights = driver.find_elements(
            By.XPATH, 
            "//div[@class='bbot-r']/h2"
        )
        for right in rights:
            if 'DOB' in right.text:
                pass
            elif 'IPL' in right.text:
                pass
            else:
                y = right.text
                lst.append(y.split(': ')[1])

        stats = driver.find_elements(
            By.XPATH, 
            "//tr[@class='ng-scope ih-td-bg']/td"
        )
        for stat in stats:
            lst.append(stat.text)
        
#         print(lst)
#         print(len(lst))
        
        try:
            data = {
                'Image':lst[0], 
                'Name':lst[1], 
                'Role':lst[2], 
                'Batting_Style': lst[3], 
                'Bowling_style': lst[4], 
                'Nationality': lst[5], 
                'Bat_Career': lst[6], 
                'Bat_matches': lst[7], 
                'Not_out': lst[8],
                'Runs': lst[9],
                'Highest_score': lst[10],
                'Avg': lst[11],
                'Ball_faced': lst[12],
                'Strike_rate': lst[13],
                '100s': lst[14],
                '50s': lst[15],
                '4s': lst[16],
                '6s': lst[17],
                'Catches': lst[18],
                'Stumping': lst[19],
                'Bowl_Career': lst[20],
                'Bowl_matches': lst[21],
                'Bowl_balls': lst[22],
                'Bowl_runs': lst[23],
                'Wickets': lst[24],
                'Best_bowlfig': lst[25],
                'Bowl_avg': lst[26],
                'Bowl_Eco': lst[27],
                'Bowl_strike_rate': lst[28],
                'Bowl_4W': lst[29],
                'Bowl_5W': lst[30],
            }
        except:
            data = {
                'Image':lst[0], 
                'Name':lst[1], 
                'Role':lst[2], 
                'Batting_Style': lst[3], 
                'Bowling_style': lst[4], 
                'Nationality': lst[5], 
                'Bat_Career': '-', 
                'Bat_matches': '-', 
                'Not_out': '-',
                'Runs': '-',
                'Highest_score': '-',
                'Avg': '-',
                'Ball_faced': '-',
                'Strike_rate': '-',
                '100s': '-',
                '50s': '-',
                '4s': '-',
                '6s': '-',
                'Catches': '-',
                'Stumping': '-',
                'Bowl_Career': '-',
                'Bowl_matches': '-',
                'Bowl_balls': '-',
                'Bowl_runs': '-',
                'Wickets': '-',
                'Best_bowlfig': '-',
                'Bowl_avg': '-',
                'Bowl_Eco': '-',
                'Bowl_strike_rate': '-',
                'Bowl_4W': '-',
                'Bowl_5W': '-',
            }
        players_details.append(data)
        print(i,' [*]Saving')

    df = pd.DataFrame(players_details)
    driver.quit()
    return df.to_csv(f'{FullTeamName}_details.csv', index = False)
```

```python
# Gujarat Titans nu baki chhe
teams = ['Chennai Super Kings', 'Delhi Capitals', 'Kolkata Knight Riders', 'Lucknow Super Giants', 'Mumbai Indians', 'Punjab Kings', 'Rajasthan Royals', 'Royal Challengers Bangalore', 'Sunrisers Hyderabad']
for team in teams:
    ipl_details(team)
```
